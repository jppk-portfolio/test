```python
## Replace non - ASCII characters
import unicodedata
def normalize_company_name(name):
    """Normalizes a company name, handling non-string values."""
    if pd.isna(name): # Check if the value is NaN or None
        return '' # Or some other placeholder
    if not isinstance(name, str): # Check if the value is a string
        try:
            name = str(name) # Attempt to convert to string
        except ValueError:
            return '' # Or some other placeholder if conversion fails
    return unicodedata.normalize('NFKD', name).encode('ASCII', 'ignore').decode()

def normalize_dataframe_column(df, column_name):
    """Applies normalization to a specified column in a DataFrame."""
    if column_name not in df.columns:
        raise ValueError(f"Column '{column_name}' not found in DataFrame.")

    df['normalized'] = df[column_name].apply(normalize_company_name)
    return df
```

```python
# remove any character that is not word or space
import pandas as pd
import unicodedata
import re

def normalize_and_clean_company_name(name):
    """Normalizes and cleans a company name, handling non-string values."""
    if pd.isna(name):
        return ''
    if not isinstance(name, str):
        try:
            name = str(name)
        except ValueError:
            return ''
    normalized_name = unicodedata.normalize('NFKD', name).encode('ASCII', 'ignore').decode()
    cleaned_name = re.sub(r'[^\w\s]', '', normalized_name)
    return cleaned_name

def normalize_and_clean_dataframe_column(df, column_name, new_column_name='normalized_cleaned'):
    """Applies normalization and cleaning to a specified column in a DataFrame."""
    if column_name not in df.columns:
        raise ValueError(f"Column '{column_name}' not found in DataFrame.")

    df[new_column_name] = df[column_name].apply(normalize_and_clean_company_name)
    return df
```

```python
def clean_name(name):
    name = name.lower().replace('"', '').replace(',', '').replace('.', '').replace('\xa0', ' ')
    suffixes = ['inc', 'ltd', 'llp', 'corp', 'corporation', 'llc', 'ulc']
    for suffix in suffixes:
        name = name.replace(f" {suffix}", "")
    return name.strip()

def process_chunk(chunk):
    chunk["Cleaned Name"] = chunk["vendor_name"].apply(clean_name)
    suggested_names = {}
    for idx, row in chunk.iterrows():
        cleaned = row["Cleaned Name"]
        found_match = False
        for key in suggested_names.keys():
            if fuzz.ratio(cleaned, key) > 85:
                suggested_names[key].append(row["vendor_name"])
                found_match = True
                break
        if not found_match:
            suggested_names[cleaned] = [row["vendor_name"]]

    chunk["Suggested Name"] = ""
    for key, names in suggested_names.items():
        suggestion = max(names, key=len)
        chunk.loc[chunk["vendor_name"].isin(names), "Suggested Name"] = suggestion.replace('"', '').strip()

    chunk = chunk.drop(columns=["Cleaned Name"])
    return chunk

def process_large_dataframe(df, chunksize=1000):
    """Processes a large DataFrame in chunks."""
    results = []
    num_chunks = len(df) // chunksize + 1 # Calculate the number of chunks

    for i in range(num_chunks):
        start = i * chunksize
        end = min((i + 1) * chunksize, len(df))
        chunk = df.iloc[start:end]
        processed_chunk = process_chunk(chunk)
        results.append(processed_chunk)

    final_df = pd.concat(results, ignore_index=True)
    return final_df
```

```python
chunked_result = process_large_dataframe(vendor_df)

# Optionally, save to a CSV file
chunked_result.to_csv("vendor_suggestions_chunked.csv", index=False)
```
